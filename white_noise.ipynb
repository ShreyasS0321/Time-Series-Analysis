```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.stattools import durbin_watson

index=[i for i in range(1000)]

data= np.random.normal(loc=0, scale=1, size=1000)

```

```python
plt.plot(index,data)

print(np.mean(data))
print(np.std(data))
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform ADF test
adf_test = adfuller(data)

print(f"ADF Statistic: {adf_test[0]}")
print(f"p-value: {adf_test[1]}")
```

```python
from statsmodels.graphics.tsaplots import plot_acf

# Plot autocorrelation
plot_acf(data, lags=40)
plt.show()
```

```python
new_data=[data[i]+data[i-1] for i in range(1,1000)]
ind=[i for i in range(len(new_data))]

```

```python
plt.plot(ind,new_data)
print(np.mean(new_data))
print(np.std(new_data))
```

```python
adf_test = adfuller(new_data)
print(f"ADF Statistic: {adf_test[0]}")
print(f"p-value: {adf_test[1]}")
```

```python
diff_dw_stat = durbin_watson(data)
print('Durbin-Watson Statistic (Differenced): {:.10f}'.format(diff_dw_stat))

diff_dw_stat = durbin_watson(new_data)
print('Durbin-Watson Statistic (Differenced): {:.10f}'.format(diff_dw_stat))
```
