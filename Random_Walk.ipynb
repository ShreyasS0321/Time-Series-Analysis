```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# Set random seed for reproducibility


# Parameters
n = 1000  # Number of steps

# Simulate random walk
steps = np.random.normal(loc=0, scale=1, size=n)
random_walk = np.cumsum(steps)
# Plot the random walk
plt.figure(figsize=(12, 6))
plt.plot(random_walk, label='Random Walk', color='blue')
plt.title('Simulated Random Walk Process')
plt.xlabel('Time')
plt.ylabel('Value')
plt.axhline(0, color='grey', lw=0.5, ls='--')
plt.legend()
plt.show()
```

```python
df = pd.DataFrame({'Random_Walk': random_walk})

# First differencing the random walk
df['Differenced'] = df['Random_Walk'].diff()

# Plot the differenced series
plt.figure(figsize=(12, 6))
plt.plot(df['Differenced'], label='Differenced Random Walk', color='purple')
plt.title('Differenced Random Walk')
plt.xlabel('Time')
plt.ylabel('Differenced Value')
plt.axhline(0, color='grey', lw=0.5, ls='--')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

adf_result = adfuller(df['Random_Walk'])
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])


adf_result = adfuller(df['Differenced'].dropna())
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
```

```python
from statsmodels.stats.stattools import durbin_watson

dw_statistic = durbin_watson(df['Random_Walk'])
print('Durbin-Watson Statistic:', dw_statistic)


dw_statistic = durbin_watson(df['Differenced'].dropna())
print('Durbin-Watson Statistic:', dw_statistic)
```
