```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline

# Step 1: Generate Synthetic Data
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(scale=0.2, size=x.shape)  # Non-linear relationship with noise

# Step 2: Fit Spline Regression
spline = UnivariateSpline(x, y, s=10)  # s is the smoothing factor
y_spline = spline(x)

# Step 3: Visualize the Data and the Spline Fit
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Synthetic Data', color='blue')
plt.plot(x, y_spline, label='Spline Fit', color='red', linewidth=2)
plt.title('Spline Regression on Synthetic Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Synthetic Data', color='blue')
#plt.plot(x, y_spline, label='Spline Fit', color='red', linewidth=2)

plt.figure(figsize=(10, 6))
#plt.plot(x, y, label='Synthetic Data', color='blue')
plt.plot(x, y_spline, label='Spline Fit', color='red', linewidth=2)
```
